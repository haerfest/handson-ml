Name the inputs X and the targets y:

There are 70,000 instances. Split it in a train set of 60,000 and a test set of 10,000:

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

```python
import numpy as np

shuffle_indexes = np.random.permutation(60000)
X_train, y_train = X_train[shuffle_indexes], y_train[shuffle_indexes]
```

The 28x28 input images contain greyscale intensities in the [0, 255] range. Scale these for the ANN to have a mean of zero:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
```

```python
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

The target class is a float (0.0-9.0) and TensorFlow will need integers, so scale them:

```python
y_train = y_train.astype(int)
y_test  = y_test.astype(int)
```

Train a Deep Neural Network with two hidden layers of 300 and 100 neurons respectively:

```python
import tensorflow as tf

feature_columns = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
```

```python
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300, 100], n_classes=10, feature_columns=feature_columns)
dnn_clf.fit(x=X_train, y=y_train, batch_size=50, steps=40000)
```

How well did it do?

# Training a DNN Using Plain Tensorflow

## Construction Phase

```python
import tensorflow as tf
import numpy as np

n_inputs  = 28*28  # MNIST
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

```python
X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
y = tf.placeholder(tf.int64, shape=(None), name='y')
```

```python
def neuron_layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2 / np.sqrt(n_inputs)
        
        # Initialize random weights.
        init = tf.truncated_normal(shape=(n_inputs, n_neurons), stddev=stddev)
        W = tf.Variable(initial_value=init, name='weights')
        
        # Initialize all biases to zero.
        b = tf.Variable(initial_value=tf.zeros([n_neurons]), name='biases')

        # Activation function.
        z = tf.matmul(X, W) + b
        if activation == 'relu':
            return tf.nn.relu(z)
        else:
            return z
```

```python
# Connect the network.
with tf.name_scope('dnn'):
    hidden1 = neuron_layer(X,       n_hidden1, 'hidden1', activation='relu')
    hidden2 = neuron_layer(hidden1, n_hidden2, 'hidden2', activation='relu')
    logits  = neuron_layer(hidden2, n_outputs, 'outputs')
```

```python
# Define the cost function (loss).
with tf.name_scope('loss'):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
    loss     = tf.reduce_mean(xentropy, name='loss')
```

```python
learning_rate = 0.01

# Define a Gradient Descent Optimizer to minimize the cost function.
with tf.name_scope('train'):
    optimizer   = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
```

```python
# Use accuracy as a measure for evaluationg the model's performance.
with tf.name_scope('eval'):
    correct  = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
init  = tf.global_variables_initializer()
saver = tf.train.Saver()
```

## Execution Phase

```python
# Use TensorFlow to load the MNIST data set.
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('/tmp/data/')
```

```python
n_epochs = 400
batch_size = 50
```

```python
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(mnist.train.num_examples // batch_size):
            X_batch, y_batch = mnist.train.next_batch(batch_size)
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch})

        acc_train = accuracy.eval(feed_dict={X: X_batch, y: y_batch})
        acc_test  = accuracy.eval(feed_dict={X: mnist.test.images, y: mnist.test.labels})
        
        print(epoch, 'Train accuracy:', acc_train, 'Test accuracy:', acc_test)
        
    save_path = saver.save(sess, './my_model_final.ckpt')
```
